```python
import matplotlib.pyplot as plt
import numpy as np

nombres = ['Red simple', 'Modelo_C3', 'Modelo_C1', 'Modelo_C2', 'Modelo_C5', 'Modelo_C4']
params = [33, 535, 931, 2883, 3363, 9859]


train_acc = [0.9679, 0.9907, 0.9943, 0.9964, 0.9900, 0.9957]
test_acc = [0.9667, 0.9700, 0.9767, 0.9700, 0.9767, 0.9767]


train_error = 1 - np.array(train_acc)
test_error = 1 - np.array(test_acc)
bias = train_error
variance = test_error - train_error


plt.figure(figsize=(12, 7))

plt.plot(nombres, bias, label='Bias (Error Train)', marker='o', color='#1f77b4', linestyle='--', linewidth=2)
plt.plot(nombres, variance, label='Variance (Gap Train-Test)', marker='o', color='#2ca02c', linestyle='--', linewidth=2)
plt.plot(nombres, test_error, label='Error Total (Test)', marker='s', color='#d62728', linewidth=4)


plt.title('Análisis de Complejidad: Bias-Variance Tradeoff', fontsize=15, pad=20)
plt.ylabel('Tasa de Error', fontsize=12)
plt.xlabel('Evolución de Arquitecturas', fontsize=12)
plt.grid(True, axis='y', linestyle=':', alpha=0.6)
plt.ylim(-0.01, 0.1)
plt.legend(fontsize=11)

plt.tight_layout()
plt.show()
```

Este gráfico muestra la evolución del Bias–Variance Tradeoff a medida que aumenta la complejidad de las arquitecturas evaluadas. Se observa que los modelos más simples presentan mayor error de entrenamiento (alto bias), mientras que al incrementar la capacidad del modelo el error total disminuye. Sin embargo, modelos excesivamente complejos no aportan mejoras significativas y pueden aumentar la varianza. El Modelo C5 representa el punto óptimo, donde se logra el mejor equilibrio entre bajo error y buena capacidad de generalización en datos no vistos.
